```python
# tak eimportant words only
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
def counter3(obj):
  L=[]
  counter = 0
  for i in ast.literal_eval(obj):
    if counter !=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(counter3)
```

```python
movies.head(
)
```

```python
def fetchdie(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i["name"])
      break
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetchdie)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# concatinate into one list before that remove space between names
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']= movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python

new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x) if isinstance(x, list) else x)
```

```python
new_df.head()
```

```python

new_df['tags'] = new_df['tags'].apply(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
# text vectorization !!!
# calculate the similarty between 2 movies , problem - textual data , approch -- getting same words thats why vectorization came in picture
# tags --> vector , nothing but a coordinate !the recommandation will be of closet vector !!
# techniquw --> bag of words : large text me se 5000 words calculate freuqncy sabse zyada -> w1 , w2 , w3 , etc
# it plot a distrubtion that calulate the number of movies and their freqent words (5000, total words) us hisab se coordinates
# will nto consider stop words like in , are , and , to , wagera
new_df['tags'] = new_df['tags'].fillna('')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# similar words (loved , loving , love to one place pe teeno)
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# we will calulate cosine distanc e--> angle between that
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
def recommand(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
  return
```

```python
recommand('Batman Begins')
#accuracy
```
